4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
import pandas as pd
orders = pd.read_csv("/content/orders.csv")
orders.head()
```

```python
orders.describe()
```

```python
orders.info()
```

```python
#Zadanie 1
orders["order_date"] = pd.to_datetime(orders["order_date"],format="%Y-%m-%d")
orders.head()
```

```python
#Zadanie 2
tshirts = orders["tshirt_category"]
tshirts.value_counts()
```

```python
#Zadanie 3 

t_shirts2 = tshirts.copy()
t_shirts2.replace("Wh Tshirt M", "White T-Shirt M", inplace=True)
t_shirts2.replace("Bl Tshirt M", "Black T-Shirt M", inplace=True)
t_shirts2.replace("Wh Tshirt F", "White T-Shirt F", inplace=True)
t_shirts2.replace("Bl Tshirt F", "Black T-Shirt F", inplace=True)
t_shirts2.value_counts()
```

```python
# Zadanie 3 metoda apply 
t_shirts2 = tshirts.copy()
t_shirts2 = t_shirts2.apply(lambda t_shirt: t_shirt.replace("Wh Tshirt M", "White T-Shirt M"))
t_shirts2 = t_shirts2.apply(lambda t_shirt: t_shirt.replace("Bl Tshirt M", "Black T-Shirt M"))
t_shirts2 = t_shirts2.apply(lambda t_shirt: t_shirt.replace("Wh Tshirt F", "White T-Shirt F"))
t_shirts2 = t_shirts2.apply(lambda t_shirt: t_shirt.replace("Bl Tshirt F", "Black T-Shirt F"))
t_shirts2.value_counts()
```

```python
#Zadanie 4
orders["tshirt_category"] = t_shirts2
orders.head()
```

```python
#4 part 1
def typ(x: str) -> str:
  if x == "Hoodie" or x == "T-Shirt":
    return x
  if x == "Tennis Shirt":
    return "other"
  _, type_, _ = tuple(x.split(" "))
  return type_
```

```python
orders["tshirt_type"] = orders["tshirt_category"].apply(lambda x: typ(x))
orders["tshirt_type"].value_counts()
```

```python
#4 part 2
def color(x: str) -> str:
  if x == "Hoodie" or x == "Tennis Shirt":
    return None
  kolor, _, _ = tuple(x.split(" "))
  return kolor
```

```python
orders["tshirt_color"] = orders["tshirt_category"].apply(lambda x: color(x))
orders["tshirt_color"].value_counts(dropna = False)
```

```python
#4 part 3
def size(x: str) -> str:
  if x == "Hoodie" or x == "Tennis Shirt":
    return None
  _, _, rozmiar = tuple(x.split(" "))
  return rozmiar
```

```python
orders["tshirt_size"] = orders["tshirt_category"].apply(lambda x: size(x))
orders["tshirt_size"].value_counts(dropna = False)
```
